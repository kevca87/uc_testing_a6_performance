```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def plot_repeat_excution(data,algorithm,ax):
    ax.plot(data, label=algorithm)
    ax.set_title(f'{algorithm}')
    ax.set_ylabel('Time (s)')
    ax.set_xlabel('Excution loop')
    return ax
```

```python
def get_algorithm_list():
    return ['bubble_sort','insertion_sort','merge_sort','quick_sort','tim_sort']
```

```python
data = pd.read_csv('algorithm_performance.csv')
```

```python
fig, ax = plt.subplots(1,5,figsize=(30,5),sharey=True)
for i, algorithm in enumerate(get_algorithm_list()):
    plot_repeat_excution(data[algorithm], algorithm, ax[i])
fig.suptitle('Sort 10 elements - Algorithms performance comparison - $timeit.repeat(number=1000,repeat=30)$')
plt.show()
```

```python
def plot_barplot_comparison(data,algorithms,ax):
    ax.bar(algorithms,data)
    ax.set_title(f'Algorithms performance comparison - $timeit.repeat(number=1000,repeat=30)$')
    ax.set_ylabel('Time (s)')
    ax.set_xlabel('Algorithm')
    return ax
```

```python
data.mean()[get_algorithm_list()].plot(kind='bar')
```

```python
fig, ax = plt.subplots(1,1)
plot_barplot_comparison(data.mean()[get_algorithm_list()],get_algorithm_list(),ax)
```

```python
data100 = pd.read_csv('algorithm_performance_100.csv')
```
